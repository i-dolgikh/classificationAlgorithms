Задачи:
- Добавить новый принак "Продолжительность"
- Удалить признаки pledget и backers, чтобы избежать утечки данных
- очистить данные от съехавших строк
- Привести категориальные признаки
- Нормализовать данные
-

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
df = pd.read_csv('data/ks-projects-201612.csv', encoding="ISO-8859-1", on_bad_lines='skip')
df.columns = df.columns.str.strip() # Избавляемся от пробелов в названияхи
```

```python
print(f"Cъехавших строк: {len(df[df['Unnamed: 13'].notna()])}") # 625
```

```python
df = df.drop(["ID", "pledged", "backers", "usd pledged", "Unnamed: 13", "Unnamed: 14", "Unnamed: 15", "Unnamed: 16"], axis=1)
df = df[df["state"].isin(["failed", "successful"])]
```

```python
df['deadline'] = pd.to_datetime(df['deadline'])
df['launched'] = pd.to_datetime(df['launched'])
df['goal'] = pd.to_numeric(df['goal'])
categorical_cols = ['category', 'main_category', 'currency', 'state', 'country']
for col in categorical_cols:
    df[col] = df[col].astype('category')
print(df.info())
```

```python
df.iloc[:6]
```

Добавляем новый признак: продолжительность компании

```python
duration = df["deadline"] - df["launched"]
df["duration"] = duration.dt.days
```

Признаки deadline и launched нам больше не нужны, удаляем их

```python
df = df.drop(["deadline", "launched"], axis=1)
```
